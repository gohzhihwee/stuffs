# **Steps #5 and #6: Performing Exploratory Data Analysis + Preliminary Insights**

The following analyses assume a US context. I have refrained from plotting non-datetime features against each other as I do not assume *a priori* that the features are automatically correlated with each other.

```python
# Standardize the time horizon of analysis before comparing datasets. Since the smallest time window we have for our data is from 2019-2022,
# we can use all data between the years 2019 and 2022
from datetime import datetime
start_date, end_date = "2019-01-01", "2022-12-31"
def standardize_timesteps(start_date, end_date, dataset):
  for i in range(len(dataset)):
    if 'Rating Date' in dataset.columns:
        dataset = dataset.rename(columns={"Rating Date": "DATE"})
    if 'Turnover' in dataset.columns:
      dataset.loc[i,'DATE'] = f"{int(dataset.loc[i,'DATE'])}-01-01"
      dataset.loc[i,'DATE'] = datetime.strptime(dataset.loc[i,'DATE'], "%Y-%M-%d")
      dataset.loc[i,'DATE'] = datetime.strftime(dataset.loc[i,'DATE'], "%Y-%M-%d")
      dataset.loc[i,'DATE'] = pd.to_datetime(dataset.loc[i,'DATE'], utc=True)
    else:
      dataset.loc[i,'DATE'] = pd.to_datetime(dataset.loc[i,'DATE'], utc=True)

  mask = (dataset['DATE'] > pd.to_datetime(start_date, utc=True)) & (dataset['DATE'] <= pd.to_datetime(end_date, utc=True))
  dataset = dataset.loc[mask]
  return dataset

auct_turn_data = standardize_timesteps(start_date, end_date, auct_turn_data)
VIX_data = standardize_timesteps(start_date, end_date, VIX_data)
etf_data = standardize_timesteps(start_date, end_date, etf_data)
cl_defa_data = standardize_timesteps(start_date, end_date, cl_defa_data)
mort_defa_data = standardize_timesteps(start_date, end_date, mort_defa_data)
adj_mortir_data = standardize_timesteps(start_date, end_date, adj_mortir_data)
cc_defa_data = standardize_timesteps(start_date, end_date, cc_defa_data)
US_GDP_data = standardize_timesteps(start_date, end_date, US_GDP_data)
cr_data = standardize_timesteps(start_date, end_date, cr_data)



```

**Scenario #1: US GDP Data vs Credit Card Loan Delinquencies from 2019-2022**

```python
cc_defa_data = cc_defa_data.merge(US_GDP_data, on='DATE', how='right').fillna(0)
fig, ax = plt.subplots(2, 1, figsize=(15,15))
ax[0].scatter(cc_defa_data['DATE'], cc_defa_data['GDPC1'])
ax[1].scatter(cc_defa_data['DATE'], cc_defa_data['DRCCLACBS'])
ax[0].set_xlabel('Year')
ax[1].set_xlabel('Year')
ax[0].set_ylabel('Real GDP (billions of 2017-chartered dollars)')
ax[1].set_ylabel('Percentage of Credit Card Loans that were Delinquent (%)')
plt.show()
```

We note the presence of similar trends in both features from the years 2019-2022 (roughly the COVID-19 pandemic years), with a drop in real GDP being followed closely by a corresponding drop in the percentage of credit card loan delinquencies. This appears to contradict the earlier hypothesis that a general negative correlation exists between economic growth (measured by real GDP) and percentage of credit card loan delinquencies as in this case, a drop in GDP was met with a drop in credit card loan delinquencies. We believe that this observation can be validated and justified through conducting further quantitative investigation into consumer spending habits during these years, as well as how external factors such as interest rate changes affected consumption and the frequency of both credit card loan issuances and credit card loan delinquencies.

In a similar vein to the observation outlined in Scenario #1, we notice an apparent positive correlation between real GDP growth in the US and the percentage of mortgage loan delinquencies, as well as a positive correlation between average mortage loan floating interest rate and the percentage of mortgage loan delinquencies during this period. The former observation warrants further investigation into both consumer spending habits as well as trends in the demand for housing and real estate during the pandemic years, while the latter is not entirely surprising considering that borrowers would find it more difficult to meet loan amounts with higher interest rates, ceteris paribus. However, we believe that more data points can be collated from alternative sources to validate this relationship as well.

**Scenario #5: US GDP Data vs Total Financial Assets held in Taxable Bond Funds/Transactions from 2019-2022**

```python
cr_data = cr_data.merge(US_GDP_data, on='DATE', how='right').fillna(0)
fig, ax = plt.subplots(2, 1, figsize=(15,15))
ax[0].scatter(cr_data['DATE'], cr_data['GDPC1'])
ax[1].scatter(cr_data['DATE'], cr_data['BOGZ1FU654091303A'])
ax[0].set_xlabel('Year')
ax[1].set_xlabel('Year')
ax[0].set_ylabel('Real GDP (billions of 2017-chartered dollars)')
ax[1].set_ylabel('Total Financial Assets held in Taxable Bond Funds/Transactions (millions of dollars)')
plt.show()
```

Unlike the strong correlation demonstrated in Scenario #4, there appears to be little correlation between the total amount of financial assets held in the form of bond funds (stored and exchanged) and real GDP growth. This would indicate that the amount of financial assets being circulated in the form of bond funds and transactions is a poor proxy measure of real GDP and hence, the probability of borrower default on a bond cannot be reliably calculated and estimated through monitoring changes in domestic real GDP growth.

**Scenario #6: VIX Scores vs Global Annual Auction Turnover from 2019-2022**

```python
VIX_data = VIX_data.merge(auct_turn_data, on='DATE', how='right').fillna(0)
for i in range(len(VIX_data)):
  if VIX_data.loc[i, "VIXCLS"] == ".":
    VIX_data.loc[i, "VIXCLS"] = 0
print(VIX_data)
fig, ax = plt.subplots(2, 1, figsize=(15,15))
ax[0].scatter(VIX_data['DATE'], VIX_data['VIXCLS'])
ax[1].scatter(VIX_data['DATE'], VIX_data['Turnover'])
ax[0].set_xlabel('Year')
ax[1].set_xlabel('Year')
ax[0].set_ylabel('VIX')
ax[1].set_ylabel('Global Annual Auction Turnover for Drawn Art (millions of dollars)')
plt.show()
```

We note that for the limited period of 2019-2022, there appears to be no correlation that can be drawn between the volatility index (VIX) and the amount spent on art as an illiquid security, as measured by auction turnover for drawn works. This suggests that the probability of borrower default on a loan for an illiquid security cannot be estimated via either the VIX or the changes in the amount spent on art globally. However, we believe that a more nuanced picture may emerge with the addition of more data points where available.

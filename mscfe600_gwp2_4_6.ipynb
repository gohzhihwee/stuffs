# **Steps #5 and #6: Performing Exploratory Data Analysis + Preliminary Insights**

The following analyses assume a US context. As in Group Work Project 1, I have refrained from plotting non-datetime features against each other as I am not assuming prior correlation.

**IMPORTANT NOTE: I HAVE INCLUDED CODE TO ADJUST THE SIZES OF THE GRAPHS AND THE X AND Y-AXES LABELS. PLEASE UNCOMMENT THE COMMENTED-OUT SETTINGS AND RERUN THE NOTEBOOK IF THE PRESENT SETTINGS RESULT IN THE GRAPH BEING TOO SMALL**

````python
# Standardize the time horizon of analysis before comparing datasets. Since the smallest time window we have for our data is from 2019-2022,
# we can use all data between the years 2019 and 2022
from datetime import datetime
start_date, end_date = "2019-01-01", "2022-12-31"
def standardize_timesteps(start_date, end_date, dataset):
  for i in range(len(dataset)):
    if 'Rating Date' in dataset.columns:
        dataset = dataset.rename(columns={"Rating Date": "DATE"})
    if 'Volume' in dataset.columns or 'Turnover' in dataset.columns:
      dataset.loc[i,'DATE'] = f"{int(dataset.loc[i,'DATE'])}-01-01"
      dataset.loc[i,'DATE'] = datetime.strptime(dataset.loc[i,'DATE'], "%Y-%M-%d")
      dataset.loc[i,'DATE'] = datetime.strftime(dataset.loc[i,'DATE'], "%Y-%M-%d")
      dataset.loc[i,'DATE'] = pd.to_datetime(dataset.loc[i,'DATE'], utc=True)
    else:
      dataset.loc[i,'DATE'] = pd.to_datetime(dataset.loc[i,'DATE'], utc=True)

  mask = (dataset['DATE'] > pd.to_datetime(start_date, utc=True)) & (dataset['DATE'] <= pd.to_datetime(end_date, utc=True))
  dataset = dataset.loc[mask]
  return dataset

US_GDP_data = standardize_timesteps(start_date, end_date, US_GDP_data)
min_pay_data = standardize_timesteps(start_date, end_date, min_pay_data)
real_hprice_data = standardize_timesteps(start_date, end_date, real_hprice_data)
con_loans_data = standardize_timesteps(start_date, end_date, con_loans_data)
margin_loan_data = standardize_timesteps(start_date, end_date, margin_loan_data)
liq_mup_data = standardize_timesteps(start_date, end_date, liq_mup_data)
liq_up_data = standardize_timesteps(start_date, end_date, liq_up_data)
liq_down_data = standardize_timesteps(start_date, end_date, liq_down_data)
liq_mdown_data = standardize_timesteps(start_date, end_date, liq_mdown_data)
dis_up_data = standardize_timesteps(start_date, end_date, dis_up_data)
dis_down_data = standardize_timesteps(start_date, end_date, dis_down_data)
dis_mdown_data = standardize_timesteps(start_date, end_date, dis_mdown_data)
art_sales_data = standardize_timesteps(start_date, end_date, art_sales_data)
````

````python
# Let's bring in the datasets used in Group Work Project 1 as well
cc_defa_data = pd.read_csv("DRCCLACBS.csv")
mort_defa_data = pd.read_csv("DRSREACBS.csv")
adj_mortir_data = pd.read_csv("MORTGAGE5US.csv")
cl_defa_data = pd.read_csv("SUBLPDRCSCLGNQ.csv")
etf_data = pd.read_csv("BOGZ1LM564091600Q.csv")
cr_data = pd.read_csv("BOGZ1FU654091303A.csv")
VIX_data = pd.read_csv("VIXCLS.csv")
auct_turn_data = pd.DataFrame(np.array([[2000,0.5], [2001,0.55], [2002,0.4], [2003,0.45], [2004,0.6], [2005,0.75], [2006,0.85], [2007,1.3], [2008,1.35], [2009,1.3], [2010,3.75], [2011,5.25], [2012,4.25], [2013,4.8], [2014,5], [2015,3.8], [2016,3.85], [2017,4.2], [2018,3.5], [2019,3.05], [2020,2.275], [2021,2.95], [2022,1.75]]), columns=['DATE', 'Turnover'])

auct_turn_data = standardize_timesteps(start_date, end_date, auct_turn_data)
VIX_data = standardize_timesteps(start_date, end_date, VIX_data)
etf_data = standardize_timesteps(start_date, end_date, etf_data)
cl_defa_data = standardize_timesteps(start_date, end_date, cl_defa_data)
mort_defa_data = standardize_timesteps(start_date, end_date, mort_defa_data)
adj_mortir_data = standardize_timesteps(start_date, end_date, adj_mortir_data)
cc_defa_data = standardize_timesteps(start_date, end_date, cc_defa_data)
cr_data = standardize_timesteps(start_date, end_date, cr_data)
````

**Scenario #1: US GDP Data vs Credit Card Loan Delinquencies from 2019-2022 (Contd.) - Who's been paying the bare minimum?**

````python
cc_defa_data = pd.merge(pd.merge(cc_defa_data,US_GDP_data,on='DATE'),min_pay_data,on='DATE').fillna(0)
fig, ax = plt.subplots(3, 1, figsize=(25,25), sharex=True)
# fig, ax = plt.subplots(3, 1, figsize=(15,15))
ax[0].scatter(cc_defa_data['DATE'], cc_defa_data['GDPC1'])
ax[1].scatter(cc_defa_data['DATE'], cc_defa_data['DRCCLACBS'])
ax[2].scatter(cc_defa_data['DATE'], cc_defa_data['RCCCBSHRMIN'])
plt.xlabel('Year')
# ax[0].set_xlabel('Year')
# ax[1].set_xlabel('Year')
# ax[2].set_xlabel('Year')
ax[0].set_ylabel('Real GDP (billions of 2017-chartered dollars)', fontsize=8)
ax[1].set_ylabel('Percentage of Credit Card Loans that were Delinquent (%)', fontsize=8)
ax[2].set_ylabel('Percentage of Accounts making the Minimum Payment on Credit Card Loans (%)', fontsize=8)
plt.show()
````

Building on our previous analysis in GWP #1, we note that the number of credit card customers paying the minimum required amount on their bills follows a similar pattern to the real GDP growth during the 2019-2022 period. The general picture appears to be that consumption was high during this period, possibly fueled by confidence (a bullish outlook) in general economic growth in the US and hence a proliferation of customers taking on credit cards, including customers who could just meet minimum payment requirements on credit card loans. Assuming that all customers comprising this statistic had minimum credit standings required, the data suggests that consumer confidence in economic outlook can lead to more customers with a wider range of credit standings applying for credit card loans, increasing the average leverage on a credit card loan as some of the customers would eventually go on to finance their loans with borrowed funds. The probability of default for a customer taking on a credit card loan at any given credit vendor would also have increased.
